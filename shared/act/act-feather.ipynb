## Session 2 Hands-on 1: Compiling GPT-OSS 20B using generated ACT-FEATHER compiler

### QKV Attention Kernel

Core component of a transformer is the QKV (Query-Key-Value) projection.

$QKV = softmax(Q \times K^T) \times V$

For GPT-OSS 20B, the dimensions of Q, K, V are `[6144, 128]`.

`attention_full.hlo`: XLA-HLO IR for the QKV kernel.  
`attention_q_sliced.hlo`: XLA-HLO IR for the QKV kernel with sliced Q matrix (6144 -> 256).

In this hands-on, we will compile the `attention_q_sliced.hlo` using the ACT-FEATHER compiler, and analyze the stages of compilation.


### Compiling with ACT-FEATHER

`cd ~/act && act-feather --input attention_q_sliced.hlo --log log/`
